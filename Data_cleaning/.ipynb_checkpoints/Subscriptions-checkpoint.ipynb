```python
import pandas as pd

# Loading Customers(Excel) data
subscriptions = pd.read_csv(r"C:\Users\Ravi Kumar\Downloads\Telecom_jupyter\Original data set\subscriptions.csv")
display(subscriptions)
```

```python
# Identifying Duplicated rows 
print("Duplicated Values -",subscriptions.duplicated().sum())
```

```python
# Cross checking column names,finding null values,improper text,Changing data types,Distinct Values 
print(subscriptions.info())
```

```python
# Changing data type of StartDate Column
subscriptions['StartDate'] = pd.to_datetime(subscriptions['StartDate'],format='%d-%m-%Y')
print(subscriptions['StartDate'])
```

```python
# Changing data type of EndDate Column
# No Need to fill null values because only Churned customers will have end dates
subscriptions['EndDate'] = pd.to_datetime(subscriptions['EndDate'],format='%d-%m-%Y')
print(subscriptions['EndDate'])
```

```python
#Checking Status column
print(subscriptions['Status'].unique())
```

```python
#Checking AutoRenew column
print(subscriptions['AutoRenew'].unique())
```

```python
#converting Total cleaned data set into CSV
subscriptions.to_csv("subscriptions_cleaned_final.csv", index=False)
```

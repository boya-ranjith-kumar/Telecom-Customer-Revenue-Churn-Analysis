```python
import pandas as pd

# Loading Customers(Excel) data
customers = pd.read_excel(r"C:\Users\Ravi Kumar\Downloads\Telecom_jupyter\data\customers.xlsx")
display(customers)
```

```python
# Identifying Duplicated rows 
print("Duplicated Values -",customers.duplicated().sum())
```

```python
# Cross checking column names,finding null values,improper text,Changing data types,Distinct Values 
print(customers.info())
```

```python
# Found null values in Age column and those are string
print(customers['age'].unique())
```

```python
# 
customers['age']=customers['age'].fillna('0')
customers['age'] = customers['age'].replace({
    'thirty': '30',
    'Twenty three': '23',
    'twenty one': '21',
    'one eight': '18',
    'five five': '55'
})
```

```python
print(customers['age'].unique())
```

```python
#Changing Data type of Age column
customers['age']=customers['age'].astype(int)
print(customers.info())
```

```python
#Filling Customers Age who contains 0 as age with mean value of that column
print(round(customers['age'].mean()))
```

```python
customers['age']=customers['age'].replace(0,round(customers['age'].mean()))
print(customers['age'].unique())
```

```python
# Checking Gender column
print(customers['Gender'].unique())
```

```python
#Correcting Gender data
customers['Gender']=customers['Gender'].replace({
    'MALE':'Male',
    'FEMALE' :'Female',
    'FEmal' :'Female'
})
print(customers['Gender'].unique())
```

```python
# Checking Region column
print(customers['Region'].unique())
```

```python
#Correcting Gender data
customers['Region']=customers['Region'].replace({
    'NORTH':'North',
    'SOUTH' :'South',
    'WEST' :'West',
    'EAST' :'East'
})
print(customers['Region'].unique())
```

```python
#Checking Joining date column
print(customers['JoinDate'])
```

```python
# Format is correct but i have to change data type
customers['JoinDate'] = pd.to_datetime(customers['JoinDate'])
print(customers['JoinDate'])
```

```python
#Checking CustomerSegment column
print(customers['CustomerSegment'].unique())
```

```python
#Checking TenureMonths column
print(customers['TenureMonths'].unique())
```

```python
#Checking DeviceType column
print(customers['DeviceType'].unique())
```

```python
#Checking SupportTicketsCount column
print(customers['SupportTicketsCount'].unique())
```
